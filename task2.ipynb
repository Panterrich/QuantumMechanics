$$ \psi_p = \frac{1}{\sqrt{2 \pi \hbar}} \int \limits_{-\infty}^{+\infty} \psi(x) \cdot e^{-\frac{i p x}{\hbar}} dx = \frac{1}{\sqrt{2 \pi \hbar}} \int \limits_{0}^{a} \sqrt{\frac{2}{a}} \sin \left(\frac{\pi x}{a} \right) \cdot e^{-\frac{i p x}{\hbar}} dx = \frac{1}{\sqrt{ \pi a \hbar}} \frac{\pi a \hbar^2 (1 + e^{iap/\hbar})}{\pi^2 \hbar^2 - a^2 p^2} $$

$$ |\psi_p|^2 = \frac{2 \pi a \hbar^3 \cos(ap/{2 \hbar})}{(\pi^2 \hbar^2 - a^2 p^2)^2} $$

```python
import pandas as pd
import sympy as smp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import math
from scipy.optimize import curve_fit
from matplotlib.animation import FuncAnimation

def psi_p(p, a = 1, h = 1):
    return 2 * np.pi * a * h ** 3 * np.abs(np.cos(a * p / 2 / h)) / (np.pi ** 2 * h ** 2 - a ** 2 * p ** 2) ** 2

def integration_simson(y, h):
    total = y[0] + y[-1]

    for i in range(1, y.size - 1):
        if (i % 2 == 1): 
            total += 4 * y[i]
        else: 
            total += 2 * y[i]

    return total * h / 3

mpl.rcParams['font.size'] = 16                   #  ,    -   
plt.figure(figsize = (8,8), facecolor = "white") #  
#    
plt.title(r'$|\psi(p)|^2$')
plt.ylabel(r'$|\psi(p)|^2$')
plt.xlabel(r"p")

p = np.linspace(-10, 10, 5000)
y = psi_p(p)
plt.plot(p, y)

plt.grid(visible = True, which = 'major', axis = 'both', alpha = 1, linewidth = 0.9)   #  
plt.grid(visible = True, which = 'minor', axis = 'both', alpha = 0.5, linestyle = ':')

plt.minorticks_on()
plt.tight_layout()
# plt.legend(loc='best', fontsize = 12) #   
plt.show()
```
